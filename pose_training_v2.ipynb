```python
X_mat1 = np.vstack( (df_concat.normalized_keypoints))
X_mat2 = np.matrix(df_concat[['angle_at_leftelbow','angle_at_rightelbow','angle_at_leftshoulder','angle_at_rightshoulder','angle_at_leftknee','angle_at_rightknee','angle_at_hip']])
X_mat = np.hstack((X_mat1,X_mat2))
#X_mat = X_mat1
y= df_concat.play
X_mat.shape
```

```python
gss = GroupShuffleSplit(n_splits=10, test_size=0.2, random_state=13)
data_split = gss.split(X_mat, y, groups = df_concat['clip'])
train_ids, test_ids = next(data_split)
```

```python

X_train, X_test, y_train, y_test = X_mat[train_ids],X_mat[test_ids],np.array(y)[train_ids],np.array(y)[test_ids] 
```

```python
clf = rfc()
clf = gbc()
#clf = LinearSVC()
#clf  = LogisticRegression()
#clf = GBC()
clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
```
